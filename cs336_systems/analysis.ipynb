```python
import pandas as pd
```

```python
def estimate_forward_mem(B, L, D):
    qkv = B * L * D * 3
    attn = B * L * L
    softmax = B * L * L
    out = B * L * D
    return (qkv + attn + softmax + out) * 4

df = pd.read_csv('../attention_stats.csv')
df['estimated_forward_mem'] = df.apply(lambda x: estimate_forward_mem(8, x.L, x.D), axis=1)
df['forward_ratio'] = df['forward_mem'] / df['estimated_forward_mem']
df['bacward_extra'] = df['backward_mem'] / df['forward_mem']
df
```
